```python
import os
import pandas as pd
import networkx as ns
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix, parallel_coordinates
import seaborn as sns
%matplotlib inline
```

```python
bos_dt = pd.read_csv('/Users/farrelmanazilin/Downloads/Dataset_M2/BostonHousing.csv')

bos_dt
```

```python
bos_dt.columns
```

```python
bos_dt = bos_dt.rename(columns={"CAT. MEDV" : "CAT_MEDV"})
bos_dt
```

```python
bos_dt.plot.scatter(x="LSTAT", y="NOX", ignore="false")
plt.show()
```

```python
fig, ax = plt.subplot()
fig.set_sizePinches(5,3)

ax.scatter
```

```python
ax = bos_dt.groupby("CHAS").mean().MEDV.plot(kind='bar', figsize= [5,3])
ax.set_ylabel("AVS MEDV")

plt.tight_layout()
plt.show()
```

```python
dataplot = bos_dt.groupby("CHAS").mean().MEDV
fig, ax = plt.subplots()
ax.bar(dataplot.index, dataplot, color=["C5", "C1"])
ax.set_xticks((0,1))
ax.set_ylabel("AVS MEDV")

plt.tight_layout()
plt.show()
```

```python
Amtrak = pd.read_csv('/Users/farrelmanazilin/Downloads/Dataset_M2/Amtrak.csv')
Amtrak.head()
```

```python
Amtrak["Data"] = pd.to_datetime(Amtrak.Month, format='%d/%m/%Y')
ridership_ts = pd.Series(Amtrak.Ridership.values, index=Amtrak.Data)
```

```python
ridership_ts.plot(ylim = [1300,2300], legend = False, figsize=[5,3])
plt.xlabel("year")
plt.ylabel("ridership")

plt.tight_layout()
plt.show()
```

```python
plt.plot(ridership_ts.index, ridership_ts)
plt.xlabel("year")
plt.ylabel("ridership")

plt.tight_layout()
plt.show()
```
